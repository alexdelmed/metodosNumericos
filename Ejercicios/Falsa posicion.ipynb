```python
#Funcion para problema de paracaidista
import math
import numpy as np
import matplotlib.pyplot as plt
def fun(c):
    return (667.38/c)*(1-math.exp(-6.81*c))-40

#Elegir los valores iniciales x0 y x1
xarray = np.linspace(10,25,100)
yarray = np.zeros(100)

for i in range(100):
    yarray[i] = fun(xarray[i])

x0 = 10
x1 = 15

for i in range(100):
    f0 = fun(x0)
    f1 = fun(x1)
    if f0*f1 > 0:
        print("No hay raiz en este rango")
        break
    x_medio = x0 + (fun(x0)(x1-x0))/(f(x1)-f(x0))
    fx_medio = fun(x_medio)
    if fx_medio*f1 < 0:
        x0 = x_medio
    else:
        x1 = x_medio
    if abs(fx_medio) < 0.001:
        print("La raiz es %.5f", x0)
        break

plt.plot(xarray, yarray)
plt.grid()
```
